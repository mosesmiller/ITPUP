These are Python's build in functions. <br>
__[Here is a link to the full reference](https://docs.python.org/3/library/functions.html)__


```python
def square (x):
    return x*x
```

```python
MyNumber = 2
result = square(MyNumber)
print(result)
```

```python
def sum_of_squares (x,y):
    return square(x) + square(y)
```

```python
number1 = 2
number2 = 3
result = sum_of_squares(number1,number2)
print(result)
```

```python
print(sum_of_squares(2,3))
```

```python
f = square
print(4, f(4))
```

```python
def fxy (f,x,y):
    return f(x) + f(y)

z = fxy(square,5,4)
print(z)
```

```python
x = 0
y = 0
def incr(x):
    y = x + 1
    return y
print(incr(5))
print(x, y)
```

```python
pi = 3.14
def area(r):
    return pi * r * r

area(1)
```

```python
numcalls = 0
def square(x):
    global numcalls
    numcalls = numcalls + 1
    return x * x
```

```python
print(square(3))
print(numcalls)
print(square(4))
print(numcalls)
```

```python
numcalls = 0
def square(x,index):
    return x * x, index + 1 
```

```python
result, numcalls = square(4,numcalls)
```

```python
print(result)
print(numcalls)
```
